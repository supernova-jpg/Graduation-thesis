```python
import cv2  
import numpy as np  
from matplotlib import pyplot as plt  
 
level = 4
def sameSize(img1, img2):   
    rows, cols, dpt = img2.shape  
    dst = img1[:rows,:cols]  
    return dst  
 
 
visible = cv2.imread("./train/vi-modify/01602D.png")  
infrared = cv2.imread("./train/ir/01602D.png")  
 
G = visible.copy()
R = infrared.copy()  
gp_visible = [G]  # Gauss Pyramid 
gp_infrared = [R]

for i in range(level):  
    G = cv2.pyrDown(G)  
    R = cv2.pyrDown(R) 
    gp_visible.append(G)  
    gp_infrared.append(R)


lp_visible = [gp_visible[level-1]]  # Laplacian pyramid
lp_infrared = [gp_infrared[level-1]]

for i in range(level-1,0,-1):  
    GE = cv2.pyrUp(gp_visible[i]) 
    GE2 = cv2.pyrUp(gp_infrared[i]) 
    L = cv2.subtract(gp_visible[i-1], sameSize(GE,gp_visible[i-1]))  
    L2 = cv2.subtract(gp_infrared[i-1], sameSize(GE2,gp_infrared[i-1])) 
    cv2.imshow('Visible Laplacian layer'+str(i),L) 
    cv2.imshow('Infrared Laplacian layer'+str(i),L2)     
    lp_visible.append(L)  
    lp_infrared.append(L2)


def Reconstruct(pyramid):
    layer = pyramid[0]

    for i in range(1,level):  
        layer = cv2.pyrUp(layer)  
        layer = cv2.addWeighted(sameSize(layer, pyramid[i]),1,pyramid[i],2,0)  
    return layer
    
     

r,c,depth = visible.shape  

 
cv2.imshow('visible',visible) 
cv2.imshow('real',infrared)  

visible_reconstruct = Reconstruct(lp_visible)
infrared_reconstruct = Reconstruct(lp_infrared)

cv2.imshow('Visible_reconstruct',visible_reconstruct)  
cv2.imshow('Infrared_reconstruct',infrared_reconstruct)  


cv2.waitKey()
cv2.destroyAllWindows()
```

### The fusion process with different layer

```python
fused_layer = []
weights = [0.6,0.4,0.2,0.1]

for i in range(level):
    layer = cv2.addWeighted(lp_visible[i], weights[i], lp_infrared[i],1-weights[i],0)
    fused_layer.append(layer)

Fused_reconstruct = Reconstruct(fused_layer)
cv2.imshow('Fused_reconstruct',Fused_reconstruct)  
plt.show()
cv2.waitKey()
cv2.destroyAllWindows()
```
